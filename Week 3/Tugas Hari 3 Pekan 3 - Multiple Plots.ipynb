<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,sharey=True,figsize=(20,8))
l1 = ax1.plot(x,tan)
ax1.set_title('Grafik tanx')
l2 = ax2.plot(x,sin,color='r')
ax2.set_title('Grafik sinx')
l3 = ax3.plot(x,cos,color='g')
ax3.set_title('Grafik cosx')

fig.legend([l1,l2,l3],labels=['tanx','cosx','sinx'], loc='upper center')
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig,ax = plt.subplots(figsize=(16,8))
ax.plot(x,y)

ins1 = ax.inset_axes([-0.1,-0.1, 0.3, 0.3])
ins1.plot(x,y2,color='red')
ins1.plot(x,y,color='blue')
ins1.set_title('Grafik CosX^2')

ins2 = ax.inset_axes([0.9,-0.1,0.3,0.3])
ins2.plot(x,y3,color='green')
ins2.plot(x,y,color='blue')
ins2.set_title('Grafik CosX^3')

ins3 = ax.inset_axes([0.9,0.9,0.3,0.3])
ins3.plot(x,y3,color='black')
ins3.plot(x,y,color='blue')
ins3.set_title('Grafik CosX^4')

ins4 = ax.inset_axes([-0.1,0.9,0.3,0.3])
ins4.plot(x,y3,color='yellow')
ins4.plot(x,y,color='blue')
ins4.set_title('Grafik CosX^5')

plt.title('Grafic CosX')
plt.show()
```
